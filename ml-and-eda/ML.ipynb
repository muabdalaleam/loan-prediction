## <center><strong>Machine<span style= "color: #F87060"> Learning</span> Models Training

Here I am going to test various ML models to see how will the models perform to filter the best ones.<br>
the data I used for the training and testing is collected in the `data-cleaning/cleaning-and-preprcessing.ipynb`

```python
def model_report(y_true, y_pred):
    
    pre  = precision_score(y_true, y_pred)
    rec  = recall_score(y_true, y_pred)
    f1   = f1_score(y_true, y_pred)
    loss = log_loss(y_true, y_pred)
    acc  = accuracy_score(y_true, y_pred)
    
    print("precision:  {:3f}\nrecall:     {:3f}\nf1-score:   {:3f}\nloss:       {:3f}\naccuracy:   {:3f}".
          format(pre, rec, f1, loss, acc))
```

```python
nca             = NeighborhoodComponentsAnalysis(n_components= 20)
knn             = KNeighborsClassifier(n_neighbors= 70, algorithm= "kd_tree")
random_forest   = RandomForestClassifier(max_depth= 40, criterion= "gini", n_estimators= 20)
desision_tree   = DecisionTreeClassifier(max_depth= 16, criterion= "entropy")
ada             = AdaBoostClassifier()
gussian_nayive  = GaussianNB()
svc             = SVC()
regressior      = LogisticRegression()
pipe            = Pipeline([("nca", nca), ("knn", knn)])

models = {"KNN"                 : knn,           "KNN & NCA"      : pipe,  "Random forest": random_forest,
          "Desision tree"       : desision_tree, "Gussian nayive" : gussian_nayive, "SVC" : svc,
          "Logistic regressior" : regressior,    "Ada Boost"      : ada,
          "SGD classifier"      : SGDClassifier()}

print("*"* 50, "Classic models performance", "*" * 50, "\n")

for mode_name, model in models.items():

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"{mode_name}: ")
    model_report(y_test, y_pred)
    print("=" * 25 + "\n")
```

## <strong><center>Deep <span style= "color: #F87060">Learning</span>

```python
X = pd.concat([X_train, X_test], axis= 0)
y = pd.concat([y_train, y_test], axis= 0)
```

```python
def plot_history(history):

    fig, (ax1, ax2) = plt.subplots(1, 2)

    ax1.plot(history.history["accuracy"],     label= "Model accuracy", color= COLORS[1])
    ax1.plot(history.history["val_accuracy"], label= "Valid accuracy", color= COLORS[2])
    ax1.set_title("Model accuracy")
    ax1.set_ylabel("accuracy")
    ax1.set_xlabel("epoch")
    # ax1.set_color_cycle(theme_colors)
    
    ax2.plot(history.history["loss"],     label= "Model loss", color= COLORS[1])
    ax2.plot(history.history["val_loss"], label= "Valid loss", color= COLORS[2])
    ax2.set_title("Model loss", )
    ax2.set_ylabel("loss")
    ax2.set_xlabel("epoch")
    # ax2.set_color_cycle(theme_colors)

    ax1.legend()
    ax2.legend()
    
    ax1.spines[['right', 'top']] .set_visible(False)
    ax2.spines[['right', 'top']] .set_visible(False)
    

    fig.suptitle("Model loss & accuracy", fontsize= FONT)
    fig.set_size_inches(12, 6)
    plt.show()
```

We will use three deffrint [**activation functions**]("https://towardsdatascience.com/activation-functions-neural-networks-1cbd9f8d91d6") for this machine learning model<br><ul>
<li> <strong>RElu</strong>: is a simple activation function I will use for first & third network and here is it's theory:<br>$RElu(x)=max(0, x)$<br><br>

<li><strong>Soft plus</strong>: is like soft RElu function and here is it's theory:<br>
$SoftPlus(x)=log(1+e^x)$<br><br>

<li><strong>Sigmoid</strong>: is a "S" shaped activation function and here is it's theory:<br>
$Sigmoid(x)=\frac{1}{1+e^-x}$

I am treating the problem as regrssion problem with tensorflow so I can get better results

```python
nn_model = keras.Sequential()

nn_model.add(keras.layers.Dense(units= 20, activation= "softplus", input_shape= (20,))) # Input layer
nn_model.add(keras.layers.Dropout(0.1))

nn_model.add(keras.layers.Dense(units= 16, activation= "softplus")) # Hidden layer
nn_model.add(keras.layers.Dropout(0.2))

nn_model.add(keras.layers.Dense(units= 32, activation= "softplus")) # Hidden layer
nn_model.add(keras.layers.Dropout(0.2))

nn_model.add(keras.layers.Dense(units= 16, activation= "softplus")) # Hidden layer
nn_model.add(keras.layers.Dropout(0.15))

nn_model.add(keras.layers.Dense(units= 1, activation= "sigmoid")) # Output layer

nn_model.compile(optimizer= keras.optimizers.Adam(0.01), 
                              loss = "mse",
                              metrics= ["accuracy"])

history = nn_model.fit(
    X_train, y_train,
    epochs= 400,
    validation_split=0.2,
    verbose= 0)

plot_history(history)
```

```python
def model_report(y_true, y_pred):
    
    round_arr = np.vectorize(lambda x: float(int(x) + 1) if (x - int(x)) > 0 else float(int(x)))
    y_pred = round_arr(y_pred)
    
    pre  = precision_score(y_true, y_pred)
    rec  = recall_score(y_true, y_pred)
    f1   = f1_score(y_true, y_pred)
    loss = log_loss(y_true, y_pred)
    acc  = accuracy_score(y_true, y_pred)
    
    print("precision:  {:3f}\nrecall:     {:3f}\nf1-score:   {:3f}\nloss:       {:3f}\naccuracy:   {:3f}".
          format(pre, rec, f1, loss, acc))

print("Nueral Network Model:")
model_report(y_test, nn_model.predict(X_test, verbose= 0))
```

## <strong><center>Combining & Visuializing the <span style= "color: #F87060">Models</span>

Now we will need to combine our models with other those haven't been trained on the `credit history`<br>
feature because as we saw in the EDA notebook thee credit history is really correlated to the loan status.

```python
X_test_combined  = X_test.copy()
X_train_combined = X_train.copy()

X_test_combined  [['Credit_History_Good', 'Credit_History_Bad']] = 1
X_train_combined [['Credit_History_Good', 'Credit_History_Bad']] = 1
```

```python
def plot_accuracy(X_test: pd.DataFrame, y_test: np.array,
                  y_pred: np.array):
    
    custom_cmap = ListedColormap([COLORS[2], COLORS[1]])
    fig, axes = plt.subplots(2, 2, sharex= True, sharey= True)

    test_data = X_test.copy()
    test_data['Loan_Status'] = y_test
    
    pred_data = X_test.copy()
    pred_data['Loan_Status'] = y_pred
    
    axes[0, 0].scatter(y= test_data[test_data['Credit_History_Good'] == 1.]['Monthly_Income'],
                       x= test_data[test_data['Credit_History_Good'] == 1.]['Loan_Amount'],
                       c= test_data[test_data['Credit_History_Good'] == 1.]['Loan_Status'],
                       cmap= custom_cmap)
    
    axes[0, 1].scatter(y= test_data[test_data['Credit_History_Good'] == 0.]['Monthly_Income'],
                       x= test_data[test_data['Credit_History_Good'] == 0.]['Loan_Amount'],
                       c= test_data[test_data['Credit_History_Good'] == 0.]['Loan_Status'],
                       cmap= custom_cmap)
    
    axes[0, 0].set_ylabel('Ground Truth')
    
    axes[0, 0].set_title('Good Credit History')
    axes[0, 1].set_title('Bad Credit History')
    
    axes[1, 0].scatter(y= pred_data[pred_data['Credit_History_Good'] == 1.]['Monthly_Income'],
                       x= pred_data[pred_data['Credit_History_Good'] == 1.]['Loan_Amount'],
                       c= pred_data[pred_data['Credit_History_Good'] == 1.]['Loan_Status'],
                       cmap= custom_cmap)
    
    axes[1, 1].scatter(y= pred_data[pred_data['Credit_History_Good'] == 0.]['Monthly_Income'],
                       x= pred_data[pred_data['Credit_History_Good'] == 0.]['Loan_Amount'],
                       c= pred_data[pred_data['Credit_History_Good'] == 0.]['Loan_Status'],
                       cmap= custom_cmap)
    
    axes[1, 0].set_ylabel('Prediction')
    
    for axs in axes:
        for ax in axs:
            ax.spines['right'] .set_visible(False)
            ax.spines['top']   .set_visible(False)

    fig.set_size_inches(10, 5)
    fig.suptitle('Credit History Effect on the Prediction', fontsize= FONT)

    plt.tight_layout()
    plt.show()
```

#### **Final Testing & Saving**

**Ada Boost & SVM**

```python
ada_1 = AdaBoostClassifier()
ada_2 = AdaBoostClassifier()

ada_1.fit(X_train, y_train)
ada_2.fit(X_train_combined, y_train)

estimators = [('ada_1', ada_1), ('ada_2', ada_2)]
ada_svc = StackingClassifier(estimators= estimators,
                             final_estimator= SVC())

ada_svc.fit(X_train, y_train)
y_pred = ada_svc.predict(X_test)

plot_accuracy(X_test, y_test, y_pred)
```

**SGD & Random forest**

```python
sgd_1 = SGDClassifier()
sgd_2 = SGDClassifier()

sgd_1.fit(X_train, y_train)
sgd_2.fit(X_train_combined, y_train)

estimators = [('sgd_1', sgd_1), ('sgd_2', sgd_2)]
sgd_rf = StackingClassifier(estimators= estimators,
                            final_estimator= RandomForestClassifier(max_depth= 50,
                                                                   criterion= "gini",
                                                                   n_estimators= 20))

sgd_rf.fit(X_train, y_train)
y_pred = sgd_rf.predict(X_test)

# model_report(y_test, y_pred)
plot_accuracy(X_test, y_test, y_pred)
```

**Logistic Regression & Decision tree**

```python
lr_1 = LogisticRegression()
lr_2 = LogisticRegression()

lr_1.fit(X_train, y_train)
lr_2.fit(X_train_combined, y_train)

estimators = [('lr_1', lr_1), ('lr_2', lr_2)]
lr_dt = StackingClassifier(estimators= estimators,
                           final_estimator= DecisionTreeClassifier(max_depth= 70))

lr_dt.fit(X_train, y_train)
y_pred = lr_dt.predict(X_test)

# model_report(y_test, y_pred)
plot_accuracy(X_test, y_test, y_pred)
```

**Deep Learning**

```python
initial_weights = nn_model.get_weights()
nn_model.set_weights(initial_weights)
nn_model.history = None

nn_model.fit(
    X_train_combined, y_train,
    epochs= 300,
    validation_split=0.2,
    verbose= 0)

nn_model.fit(
    X_train, y_train,
    epochs= 300,
    validation_split=0.2,
    verbose= 0)

y_pred = nn_model.predict(X_test, verbose= 0)

plot_accuracy(X_test, y_test, y_pred)
```
